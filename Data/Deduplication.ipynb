```python
# Deduplicate Sentences in a CSV File

# 1. Setup and Imports
# Import the pandas library, which is essential for working with CSV and tabular data.
import pandas as pd
import os

# --- CONFIGURATION ---
# IMPORTANT: Update these variables with your actual file names and column name.

# The name of your input CSV file (the one with duplicates).
INPUT_FILE_NAME = 'Laptop_Train_v2.csv'

# The name for the new CSV file that will contain only unique sentences.
OUTPUT_FILE_NAME = 'Laptop_updated.csv'

# The name of the column that contains the sentences (e.g., 'Sentence', 'ReviewText', 'Body').
# This column is case-sensitive, so ensure it matches exactly!
SENTENCE_COLUMN_NAME = 'Sentence'

# -----------------------

# 2. Function to Deduplicate and Save
def deduplicate_and_save(input_path, output_path, column_name):
    """
    Reads a CSV, finds unique values in a specified column, and saves
    only the rows with unique values to a new CSV file.
    """
    try:
        # Load the CSV file into a pandas DataFrame.
        print(f"Loading data from: {input_path}")
        df = pd.read_csv(input_path)

        # Check if the required column exists in the DataFrame
        if column_name not in df.columns:
            print(f"\nERROR: Column '{column_name}' not found in the input file.")
            print(f"Available columns are: {list(df.columns)}")
            return

        # Get the initial count of rows
        initial_count = len(df)

        # Remove duplicate rows based ONLY on the specified sentence column.
        # `subset=[column_name]` tells pandas to only consider that column for duplication checks.
        # `keep='first'` ensures the first occurrence of a unique sentence is kept.
        df_unique = df.drop_duplicates(subset=[column_name], keep='first')

        # Get the final count of unique rows
        unique_count = len(df_unique)

        # Save the unique DataFrame to a new CSV file
        df_unique.to_csv(output_path, index=False)

        # Print summary statistics
        print("\n--- Processing Complete ---")
        print(f"Initial row count: {initial_count}")
        print(f"Unique row count: {unique_count}")
        print(f"Number of duplicates removed: {initial_count - unique_count}")
        print(f"Data saved successfully to: {output_path}")

    except FileNotFoundError:
        print(f"\nERROR: The file '{input_path}' was not found. Please check the file name and path.")
    except Exception as e:
        print(f"\nAn unexpected error occurred: {e}")

# 3. Execution
# Note: You should place your 'input_reviews.csv' file in the same directory as this notebook
# or provide the full path in the INPUT_FILE_NAME variable.
deduplicate_and_save(INPUT_FILE_NAME, OUTPUT_FILE_NAME, SENTENCE_COLUMN_NAME)
```
