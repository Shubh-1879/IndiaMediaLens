```python
import pandas as pd
import math
import openpyxl

# ---- CONFIG ----
input_file = "Laptop_Final_Final.csv"       # your CSV file
column_name = "Sentence"      # the column header
batch_size = 200               # sentences per batch
# ----------------

# Read the CSV file
df = pd.read_csv(input_file)

# Extract the column
sentences = df[column_name]

# Number of batches
num_batches = math.ceil(len(sentences) / batch_size)

for i in range(num_batches):
    start = i * batch_size
    end = start + batch_size
    batch_df = sentences[start:end].to_frame()

    # Output filename
    filename = f"batch {i+1}.xlsx"

    # Save each batch as its own Excel file
    batch_df.to_excel(filename, index=False)

    print(f"Created: {filename} ({len(batch_df)} rows)")
```

```python
import pandas as pd
import json

# ---- CONFIG ----
json_file = "batch8_out.json"           
output_excel = "batch8_expanded.xlsx"   
# ----------------

# Load JSON
with open(json_file, "r", encoding="utf-8") as f:
    data = json.load(f)

rows = []

# Build rows
for item in data:
    sentence = item["sentence"]
    aspects = item["aspects"]

    # if no aspects (rare case)
    if not aspects:
        rows.append({
            "sentence": sentence,
            "term": "",
            "polarity": ""
        })
    else:
        for asp in aspects:
            rows.append({
                "sentence": sentence,
                "term": asp["term"],
                "polarity": asp["polarity"]
            })

# Create DataFrame
df = pd.DataFrame(rows)

# Save Excel
df.to_excel(output_excel, index=False)

print("Done! Saved:", output_excel)
```
