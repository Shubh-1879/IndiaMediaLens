```python
import pandas as pd
import os

def create_grouped_batches(file_path, batch_size_target=200):
    # Load the dataset
    df = pd.read_excel(file_path)
    
    # Fill any missing sentences temporarily to ensure grouping works
    df['Sentence'] = df['Sentence'].fillna('UNKNOWN_SENTENCE')
    
    # Group by Sentence so we can process them as units
    grouped = df.groupby('Sentence', sort=False)
    
    current_batch_list = []
    current_batch_count = 0
    batch_number = 1
    
    for sentence, group_df in grouped:
        # If adding this entire group exceeds the target size, 
        # and we already have some data in the current batch, save it first.
        if current_batch_count + len(group_df) > batch_size_target and current_batch_list:
            # Save the current batch
            batch_df = pd.concat(current_batch_list)
            save_name = f"Laptop_Train_Raw_Batch_{batch_number}.xlsx"
            batch_df.to_excel(save_name, index=False)
            print(f"Saved: {save_name} with {len(batch_df)} rows.")
            
            # Reset for next batch
            current_batch_list = [group_df]
            current_batch_count = len(group_df)
            batch_number += 1
        else:
            # Accumulate groups into the current batch
            current_batch_list.append(group_df)
            current_batch_count += len(group_df)
            
    # Save the final batch
    if current_batch_list:
        batch_df = pd.concat(current_batch_list)
        save_name = f"Laptop_Train_Raw_Batch_{batch_number}.xlsx"
        batch_df.to_excel(save_name, index=False)
        print(f"Saved final: {save_name} with {len(batch_df)} rows.")

# Run the function
create_grouped_batches('Laptop_Train_Raw_Template.xlsx')
```
